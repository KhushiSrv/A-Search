```python
import networkx as nx
import matplotlib.pyplot as plt
import random

G = nx.Graph()
# Add nodes
G.add_node('A',weight=10)
G.add_node('B',weight=8)
G.add_node('C',weight=5)
G.add_node('D',weight=7)
G.add_node('E',weight=3)
G.add_node('F',weight=6)
G.add_node('G',weight=5)
G.add_node('H',weight=3)
G.add_node('I',weight=1)
G.add_node('J',weight=0)

# Add weighted edges
G.add_edge('A', 'B', weight=6)
G.add_edge('B', 'C', weight=3)
G.add_edge('B', 'D', weight=2)
G.add_edge('C', 'D', weight=1)
G.add_edge('C', 'E', weight=5)
G.add_edge('D', 'E', weight=8)
G.add_edge('E', 'I', weight=5)
G.add_edge('E', 'J', weight=5)
G.add_edge('J', 'I', weight=3)
G.add_edge('F', 'G', weight=1)
G.add_edge('F', 'H', weight=7)
G.add_edge('G', 'I', weight=3)
G.add_edge('H', 'I', weight=2)
G.add_edge('A', 'F', weight=3)


def astar_search(graph, start, goal):
    open_set = [(0, start, [start])]  # (f-cost, node, path)
    visited = set()
    while open_set:
        open_set.sort(key=lambda x: x[0])  # Sort the open set by f-cost (lowest first)
        current_cost, current_node, current_path = open_set.pop(0)  # Get the node with the lowest f-cost
        if current_node == goal:
            return current_cost, current_path  # Return the cost and path if the goal is reached
        if current_node in visited:
            continue
        visited.add(current_node)
        for neighbor in graph.neighbors(current_node):
            if neighbor not in visited:
                cost = current_cost + graph[current_node][neighbor]['weight']
                heuristic_cost = heuristic(neighbor, goal)  # A* heuristic (in this example, it's just 0)
                f_cost = cost + heuristic_cost
                new_path = current_path + [neighbor]
                open_set.append((f_cost, neighbor, new_path))
    return float('inf'), []  # If no path is found, return an empty list


# Define the A* heuristic function (in this example, it's 0)
def heuristic(node, goal):
    return G.nodes[node]['weight']
```

```python
# Start and goal nodes
start_node = 'A'
goal_node = 'J'

# Find the shortest path using A* search
shortest_path = astar_search(G, start_node, goal_node)
print(shortest_path)
```

```python
pos = nx.spring_layout(G,random.seed(420))

nx.draw(G, pos, with_labels=True, node_size=500, node_color='blue', font_size=12, font_color='black')
labels = nx.get_edge_attributes(G, 'weight')

edges=[]
for i in range(1,len(shortest_path[1])):
    t1=(shortest_path[1][i-1],shortest_path[1][i])
    edges.append(t1)

nx.draw_networkx_edges(G,pos,edgelist=edges,edge_color='red')
nx.draw_networkx_edge_labels(G, pos,edge_labels=labels)

plt.show()
```
